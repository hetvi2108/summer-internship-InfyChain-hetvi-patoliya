## Step 3: Encoding Categorical Variables

**Goal:** Convert all categorical columns into numerical representations
so they can be used by machine learning algorithms.

| Column | Type | Strategy | Reason |
|--------|------|----------|---------|
| `Sex` | Binary | Label Encoding | Only 2 values (male/female) |
| `Embarked` | Nominal | One-Hot Encoding | 3 unordered categories |
| `Pclass` | Ordinal | Already numeric | No change needed |

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

sns.set_theme(style='whitegrid')

df = pd.read_csv('../data/cleaned_dataset.csv')
print(f'Shape: {df.shape}')
df.head()
```

### 3.1 Identify Categorical Columns

```python
# Check for both 'object' (strings) and 'category' dtypes
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print(f'Categorical columns found: {len(cat_cols)}')
print(f'List: {cat_cols}\n')

if len(cat_cols) > 0:
    for col in cat_cols:
        # Standard spaces used for indentation here!
        print(f'{col}: {df[col].unique()}')
else:
    print("No categorical columns left! Everything is numeric.")
```

### 3.2 Verify All Columns are Numeric

```python
print('Data types after encoding:')
print(df.dtypes)
print(f'\nAny object columns remaining: {df.select_dtypes(include="object").shape[1]}')
```
