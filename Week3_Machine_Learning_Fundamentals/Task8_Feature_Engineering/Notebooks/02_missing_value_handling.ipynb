## Step 2: Missing Value Handling

**Goal:** Treat all missing values using appropriate strategies —
median imputation for numerical, mode for categorical, and
dropping columns with excessive nulls.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

# Added 'r' before the string to safely read the Windows path
df = pd.read_csv(r'E:\summer-internship-2026-InfyChain-Kinari-Thummar\Week-3_Machine_Learning\Task9_Feature_Engineering\data\titanic.csv')

print(f'Shape: {df.shape}')
df.head()
```

### 2.1 Missing Values Before Treatment

```python
before = df.isnull().sum()
before_pct = (df.isnull().sum() / len(df) * 100).round(2)
print(pd.DataFrame({'Count': before, 'Percent': before_pct})[before > 0])
```

### 2.2 Check and Remove duplicate rows

```python
duplicates_count = df.duplicated().sum()
print(f"--- Duplicate Check ---")
print(f"Duplicate rows found: {duplicates_count}")

if duplicates_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicates removed successfully.")
```

```python
df.columns
```

### 2.3 Drop Irrelevant Columns

Columns that carry no predictive signal and would only add noise.

```python
# 1. Explicitly drop the ID column (no predictive value)
cols_to_drop = ['Passengerid']

# 2. Dynamically find all columns that start with 'zero'
zero_cols = [col for col in df.columns if col.startswith('zero')]
cols_to_drop.extend(zero_cols)
```

### 2.4 Dropping and Cleaning 

```python
# Drop the identified columns
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# Fix the typo in the target column name for best practices
df.rename(columns={'2urvived': 'Survived'}, inplace=True)
```

```python
print(f"Dropped {len(cols_to_drop)} columns: {cols_to_drop}\n")
print(f"Remaining columns to work with:\n{df.columns.tolist()}")
```

### 2.5 Save Output

```python
df.to_csv('cleaned_dataset.csv', index=False)
print("Dataset successfully cleaned and saved to 'cleaned_dataset.csv'")
```
